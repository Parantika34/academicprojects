```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('DailyDelhiClimateTrain.csv', index_col='date', parse_dates=True)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.plot(subplots=True, figsize=(14,6))
plt.show()
```

```python
df_monthly = df['meantemp']
df_monthly.plot(figsize=(14,6))
plt.show()
```

```python
df_monthly = df['meantemp'].resample('MS').mean()
df_monthly.plot(figsize=(14,6))
plt.show()
```

```python
df_monthly = df['humidity'].resample('MS').mean()
df_monthly.plot(figsize=(14,6))
plt.show()
```

```python
df_monthly = df['meanpressure'].resample('MS').mean()
df_monthly.plot(figsize=(14,6))
plt.show()
```

```python
df_monthly = df['wind_speed'].resample('MS').mean()
df_monthly.plot(figsize=(14,6))
plt.show()
```
